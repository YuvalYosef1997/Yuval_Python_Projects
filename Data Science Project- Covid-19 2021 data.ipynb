# Graphs:

```python
# Graph of the total Covid-19 deaths cases as a function of each continent. 
sns.catplot(data=covid19_data,x="continent",y="total_deaths",kind="bar")
```

```python
# Sampling the all Dataset to data of some countries with a similar population size, in order to find statistics.
High_population_countries= covid19_data.query("location == ['United States','China','India','Brazil']")
Low_population_countries= covid19_data.query("location == ['Israel','Serbia','Switzerland','United Arab Emirates']")
```

```python
# Grpah which shows the total Covid-19 cases,depending on these high population countries.
sns.catplot(x='population',y='total_cases',kind="bar",hue="location",data=High_population_countries)
```

```python
# Grpah which shows the total Covid-19 cases,depending on these low population countries.
sns.catplot(x='population',y='total_cases',kind="bar",hue="location",data=Low_population_countries)
```

```python
# Grpah which shows the total Covid-19 deaths-cases,depending on these high population countries
sns.catplot(x='population',y='total_deaths',kind= "bar",hue='location',data=High_population_countries)
```

```python
# Grpah which shows the total Covid-19 deaths-cases,depending on these low population countries
sns.catplot(x='population',y='total_deaths',kind= "bar",hue='location',data=Low_population_countries)
```

```python
# Grpah which shows the total vaccinations(of Pfizer and BioNTech only),depending on these high population countries.
sns.catplot(x='population',y='total_vaccinations',kind= "bar",hue='location',data=High_population_countries)
```

```python
# Grpah which shows the total vaccinations(of Pfizer and BioNTech only),depending on these low population countries.
sns.catplot(x='population',y='total_vaccinations',kind= "bar",hue='location',data=Low_population_countries)
```

```python
# Grpah which shows the total amount of people who fully vaccinated (of Pfizer and BioNTech only),depending on these high population countries.
sns.catplot(x='population',y='people_fully_vaccinated',kind= "bar",hue='location',data=High_population_countries)
```

```python
# Grpah which shows the total amount of people who fully vaccinated (of Pfizer and BioNTech only),depending on these low population countries.
sns.catplot(x='population',y='people_fully_vaccinated',kind= "bar",hue='location',data=Low_population_countries)
```

```python
# Grpah which shows the total Covid-19 tests,depending on these high population countries.
sns.catplot(x='population',y='total_tests',kind="bar",hue="location",data=High_population_countries)
```

```python
# Grpah which shows the total Covid-19 tests,depending on these low population countries.
sns.catplot(x='population',y='total_tests',kind="bar",hue="location",data=Low_population_countries)
```

```python
vaccinations_time=covid19_data.query('people_fully_vaccinated>0')
```

```python
# Graph of the new cases of Covid-19 worldwide as a function of the people who fully vaccinated(of Pfizer and BioNTech only) worldwide.
sns.relplot(x='people_fully_vaccinated',y='new_cases',hue='continent',data=vaccinations_time)
```

```python
# Graph of the new cases of Covid-19 worldwide as a function of the people who fully vaccinated (of Pfizer and BioNTech only) worldwide,with a line-chart type.
sns.relplot(x='people_fully_vaccinated',y='new_cases',kind= 'line',palette='viridis',hue='continent',data=vaccinations_time)
```

```python
# Graph of the new deaths cases of Covid-19 worldwide as a function of the people who fully vaccinated (of Pfizer and BioNTech only) worldwide. 
sns.relplot(x='people_fully_vaccinated',y='new_deaths',hue='continent',data=vaccinations_time)
```

```python
# Graph of the new deaths cases of Covid-19 worldwide as a function of the people who fully vaccinated  (of Pfizer and BioNTech only)worldwide,with a line-chart type.
sns.relplot(x='people_fully_vaccinated',y='new_deaths',kind= 'line',hue='continent',palette='viridis',data=vaccinations_time)
```
